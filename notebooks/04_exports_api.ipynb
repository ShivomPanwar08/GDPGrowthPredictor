```python
import numpy as np
import pandas as pd
import requests
```

```python
url = 'https://api.worldbank.org/v2/country/all/indicator/NE.EXP.GNFS.ZS?format=json&per_page=20000'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
data = response.json()
```

```python
type(data)
```

```python
data
```

```python
data[0]
```

```python
data[1][0]
```

```python
records = data[1]
```

```python
df = pd.DataFrame(records)
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[['country', 'countryiso3code', 'date', 'value']]
```

```python
df.head()
```

# Data cleaning

```python
df['country'] = df['country'].apply(lambda x : x['value'])
```

```python
df.to_csv('exports.csv', index = False)
```

```python
import numpy as np
import pandas as pd
import requests
```

```python
url = 'https://api.worldbank.org/v2/country/all/indicator/NE.GDI.TOTL.ZS?format=json&per_page=20000'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
data = response.json()
```

```python
type(data)
```

```python
data[0]
```

```python
data[1][0]
```

```python
df = pd.DataFrame(data[1])
```

```python
df.head()
```

```python
df.columns
```

```python
df = df[['country', 'countryiso3code', 'date', 'value']]
```

```python
df.head()
```

```python
df['country'] = df['country'].apply(lambda x : x['value'])
```

```python
df.isnull().sum()
```

```python
df.to_csv('capital.csv', index = False)
```

```python
import numpy as np
import pandas as pd
import requests 
```

```python
url = 'https://api.worldbank.org/v2/country/all/indicator/FP.CPI.TOTL.ZG?format=json&per_page=20000'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
data = response.json()
```

```python
records = data[1]
```

```python
df = pd.DataFrame(records)
```

```python
df = df[['country', 'countryiso3code', 'date', 'value']]
```

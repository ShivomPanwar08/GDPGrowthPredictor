# Data Collection of GDP Growth Rate

```python
url = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.MKTP.KD.ZG?format=json&per_page=20000"  #It's mention that per_page must be 20000 data entries.
```

```python
response = requests.get(url)   # API Call
```

```python
response.status_code
```

```python
data = response.json()    #Converting the API into JSON.
```

```python
type(data)
```

```python
data
```

```python
data[0]
```

```python
data[1][0]
```

```python
records = data[1]
```

```python
df = pd.DataFrame(records)
```

```python
df.shape
```

```python
df.columns
```

```python
df = df[['country', 'countryiso3code', 'date', 'value']]
```

# Data Cleaning

```python
df['country'] = df['country'].apply(lambda x : x['value']) #To separate key i.e.is id here so that we get only country name.
```

```python
import numpy as np
import pandas as pd
```

```python
gdp = pd.read_csv("../data/raw/gdp_growth.csv").rename(columns={'value' : 'gdp_growth'})
cap = pd.read_csv('../data/raw/capital.csv').rename(columns={'value' : 'capital'})
exp = pd.read_csv('../data/raw/exports.csv').rename(columns={'value' : 'exports'})
fdi = pd.read_csv('../data/raw/fdi.csv').rename(columns={'value' : 'fdi'})
gov = pd.read_csv('../data/raw/gov_final.csv').rename(columns={'value' : 'govt'})
imp = pd.read_csv('../data/raw/imports.csv').rename(columns={'value' : 'imports'})
inf = pd.read_csv('../data/raw/inflation.csv').rename(columns={'value' : 'inf'})
pop = pd.read_csv('../data/raw/population.csv').rename(columns={'value' : 'pop'})
ump = pd.read_csv('../data/raw/unemployment.csv').rename(columns={'value' : 'ump'})
```

```python
# GDP base dataframe
df = gdp.copy()

# Merge remaining datasets
df = df.merge(
    inf[["country", "date", "inf"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    ump[["country", "date", "ump"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    pop[["country", "date", "pop"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    exp[["country", "date", "exports"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    imp[["country", "date", "imports"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    fdi[["country", "date", "fdi"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    cap[["country", "date", "capital"]],
    on=["country", "date"],
    how="left"
)

df = df.merge(
    gov[["country", "date", "govt"]],
    on=["country", "date"],
    how="left"
)

```

```python
import requests

# Fetch country metadata to identify aggregates
url = "http://api.worldbank.org/v2/country?format=json&per_page=400"
resp = requests.get(url).json()
countries_meta = resp[1]

meta_df = pd.DataFrame([{
    'countryiso3code': c['id'],
    'region': c['region']['value']
} for c in countries_meta])

aggregates = meta_df[meta_df['region'] == 'Aggregates']['countryiso3code'].tolist()

# Check what's in your data
found_aggregates = df[df['countryiso3code'].isin(aggregates)]['country'].unique()
print(f"Aggregates found in df ({len(found_aggregates)}):")
print(found_aggregates)

# Filter them out
df = df[~df['countryiso3code'].isin(aggregates)].reset_index(drop=True)

print(f"\nShape after filtering: {df.shape}")
print(f"Unique countries after: {df['country'].nunique()}")
```

```python
import pandas as pd

df = pd.read_csv('../data/processed/finale.csv')  # your file is actually CSV format despite the .xls name
print(f"Shape before: {df.shape}")
print(f"Unique countries before: {df['country'].nunique()}")

# These are the leftover aggregates — income groups + not classified
unwanted_categories = [
    'High income', 'Upper middle income', 'Lower middle income', 
    'Low income', 'Middle income', 'Low & middle income',
    'Not classified'
]

df_clean = df[~df['country'].isin(unwanted_categories)].reset_index(drop=True)

print(f"\nShape after: {df_clean.shape}")
print(f"Unique countries after: {df_clean['country'].nunique()}")

# Confirm they're gone
remaining_unwanted = df_clean[df_clean['country'].isin(unwanted_categories)]
print(f"\nUnwanted rows remaining (should be 0): {len(remaining_unwanted)}")

df_clean.to_csv('finale_clean.csv', index=False)
```
